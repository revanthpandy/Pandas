## Cleaning dataSet CSV (Revision)

```python
import pandas as pd
```

## Inspecting Data

```python
df.info()
```

## Cleaning Column names

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df.head()
```

## Fixing date Column


```python
date_cols = ["orderdate","ship_date"]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")
    
```

```python
df.head()
```

## Fixing Money Value

```python
money_col = ["unit_price","shippingcost","order_total"]

for col in money_col:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace("$","")
        .str.replace("USD","")
        .str.strip()
        
    )
    df[col] = pd.to_numeric(df[col],errors="coerce")
    
```

```python
df.head()
```

## Fixing percentage

```python
df["taxrate"] =(
        df["taxrate"]
        .astype(str)
        .str.replace("%","")
) 
df["taxrate"] = pd.to_numeric(df["taxrate"],errors="coerce")
df.loc[df["taxrate"]>1,"taxrate"] =df["taxrate"]/100
```

```python
df.head()
```

```python
df["quantity"] = pd.to_numeric(df["quantity"],errors="coerce")
df["quantity"] = df["quantity"].fillna(1)
```

## Handeling Missing Value


```python
df.isna().sum()
```

## Removing Dupliacates


```python
df.shape
```

```python
df = df.drop_duplicates()
```
